
# Analyze the data

- Identify the distribution of each variable
- Missing data
- Outliers
- Correlated data

```
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```
with open('../data/df_train.csv', 'rb') as handle:
    df_train = pickle.load(handle)
```

```
df_train
```

```
print('Shape : ', df_train.shape)
# Many columns compared to the number of rows ...
```

```
df_train.describe()
```

```
df_train.info()
# All data are already digitized (except music_id and category)
```

## Analyze continious variables

```
continious_variables = pd.DataFrame(df_train.columns.tolist())
continious_variables = continious_variables.iloc[1:df_train.shape[1]-1]
for col in continious_variables[0]:
    plt.figure()
    sns.displot(df_train[col], kde=True)
```

### Interpretation

The variables appear to be well distributed. They should be standardized for further comparison

## Target variable analysis

```
plt.figure()
df_train['category'].value_counts().plot.pie(autopct="%.1f%%")
plt.title('Breakdown of the Target variable')
plt.legend()
```

#### Interpretation:
The dataset is evenly distributed over the "Category" variable

## Missing Data Analysis

```
[(col + " : " + str(df_train[col].isnull().sum())) for col in df_train.columns]
```
